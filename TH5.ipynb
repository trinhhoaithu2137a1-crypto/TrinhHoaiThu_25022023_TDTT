```python
# W5A3
import math
def prime(n):
    if n <= 1:
        return False
    elif n == 2:
        return True
    else:
        for i in range(2, math.isqrt(n)+1):
            if n%i == 0:
                return False
        return True
if __name__=="__main__":
    print(prime(5))
```

```python
#W5A4
import math
def Perfect_Number(n:int):
    if n == 1:
        return False
    else:
        tong = 1
        for i in range(2, math.isqrt(n)+1):
            if n%i == 0:
                tong += i
                if i*i != n:
                    tong += n/i
        return n == tong
if __name__=="__main__":
    print(Perfect_Number(6))
```

```python
#W5A5
def find_first_occurrence(Danh_sach: list[int], k:int):
    count = 0
    for i in range(len(Danh_sach)):
        if Danh_sach[i] == k:
            count = i+1
            return count
    return -1
if __name__=="__main__":
    Danh_sach_1 = [2, 3, 5, 6]
    k_to_find = 5
    print(find_first_occurrence(Danh_sach_1, k_to_find))
```

```python
#W5A5 hàm nhập được giá trị k và list
def find_first_occurrence(k):
    try:
        input_str = input()
        if not input_str.strip():
            return -2
        data_iterator = map(int, input_str.split())
        
    except ValueError:
        return -1
    index = 0
    for number in data_iterator:
        if number == k:
            return index 
        index += 1
    return -1
if __name__ == '__main__':
    try:
        k_to_find = int(input())
        result = find_first_occurrence(k_to_find)
        if result == -2:
            print("Lỗi: Đầu vào không hợp lệ")
        elif result == -1:
            print("-1")
        else:
            print(result+1)
    except ValueError:
        print("Kí tự bạn nhập không hợp lệ")
```

```python
#W5A6
def factorial(n):
    if n<=0:
        return False
    else:
        giai_thua = 1
        while n:
            giai_thua *= n
            n -= 1
        return giai_thua
if __name__ == "__main__":
    print(factorial(4))
```

```python
#W5A8
def binary(n):
    binary_result = ""
    temp_n = n
    while temp_n > 0:
        du = temp_n % 2
        binary_result = binary_result + str(du)
        temp_n //= 2
    return binary_result
def xor(num1, num2):
    ketqua_xor = num1^num2
    return ketqua_xor
def reverse(n):
    du = 0
    while n:
        du = du*10 + n%10
        n //= 10
        return du
def hamming(a, b):
    result = xor(a, b)
    nhi_phan = int(binary(result))
    count = 0
    while nhi_phan:
        if nhi_phan % 10 == 1:
            count+=1
        nhi_phan//=10
    return count
if __name__=="__main__":
    number_1 = 1
    number_2 = 4
    print(hamming(number_1, number_2))
```

```python
#W5A9
def sum_digit(n):
    du = 0
    sum_digit = 0
    while n:
        du = n%10
        sum_digit += du
        n //= 10
    return sum_digit
if __name__=="__main__":
    so_nguyen_duong = 222222222
    print(sum_digit(so_nguyen_duong))
```

```python
#W5A10
def dang_cau(str_1: str, str_2: str):
    if len(str_1) != len(str_2):
        return False
    else:
        n = len(str_1)
        map1 = {}
        map2 = {}
        for i in range(n):
            char1 = str_1[i]
            char2 = str_2[i]
            
            if char1 in map1:
                if map1[char1] != char2:
                    return False
            else:
                map1[char1] = char2
                
            if char2 in map2:
                if map2[char2] != char1:
                    return False
            else:
                map2[char2] = char1
    
        return True
    
if __name__ == "__main__":
    map1 = "2345"
    map2 = "3456"
    print(dang_cau(map1, map2))
```
