```python
#W2A4
a1, b1, c1, a2, b2, a3= map(int,input().split())
# dùng hàm map để ép các phần tử sang kiểu dữ liệu mong muốn
# int là kiểu số nguyên
# split dùng để tách các chuỗi kí tự
diemtrungbinh = int(((a1+b1+c1)+(a2+b2)*2+a3*3))/10
print ("Điểm trung bình:%.1f" % diemtrungbinh)
```
