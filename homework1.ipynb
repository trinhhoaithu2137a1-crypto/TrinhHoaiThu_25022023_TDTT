```python
#Bai 5
c = input("Nhập chữ cái hoa:")
lower = c.lower()
before = chr(ord(lower)- 1)
if c == 'A':
    print(" A là TH đặc biệt")
elif "A" <= c <= "Z":
    print (f"Chữ cái thường liền trước:{before}")
```

```python
#Bai 6
import math as mt
a, b, c = map(float, input().split())
if a>0 and b>0 and c>0 and a+b>c and b+c>a and a+c>b:
  p=(a+b+c)/2
  s= mt.sqrt(p*(p-a)*(p-b)*(p-c))
  print(f"Dien tich hinh tam giac la:{s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
#Bai 7
s = input("Nhập vào một chuỗi kí tự thường:")
if len(s) >= 20 and s.islower(): # len(s) kiem tra do dai ki tu, s.islower kiem tra chu viet thuong
    print ("Kí tự thứ 5:", s[4])# s[4] ki tu thu 5 vi bat dau bang 0
    print ("Kí tự thứ 9:", s[8])
else:
    print (" Chuỗi kí tự không hợp lệ")
```
