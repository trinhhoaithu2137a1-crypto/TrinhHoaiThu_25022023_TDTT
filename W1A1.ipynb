```python
#W1A3
n=4 # khởi tạo biến n với giá trị số nguyên là 4
t=0 # khởi tạo biến t dùng để lưu tổng với giá trị là 0
for i in range(1, n+1): # tạo vòng lặp for, hàm range tạo ra dãy số từ 1 đến n+1 
    t+=i # cộng các giá tị hiện tại của i vào biến t
print (t) # in ra giá trị cuối cùng của t ra màn hình
```

```python
#W1A4
number = [1, 2, 3, 4] # dòng này tạo một danh sách (list) có tên number chứa các số nguyên từ 1 đến 4
for x in number: # dòng này là vòng lặp duyệt qua các phần tử trong danh sách, gán lần lượt giá trị của phần tử cho biến x
    if x % 2 == 0: # kiểm tra điều kiện phép chia lấy dư nếu x chia 2 dư 0 
        print(x, " là số chẵn") # in ra giá trị của x là số chẵn nếu chia 2 dư 0
    else: # khối lệnh thay thế sẽ được thực thi nếu if là sai
        print(x, " là số lẻ") # in ra giá trị của x là số lẻ nếu if sai
```

```python
#W1A5
animals= ["cat","dog","cat","bird"] # dòng này tạo ra một danh sách các chuỗi (strings) có tên animals
count = 0 # khởi tạo biến count với giá trị ban đầu =0
for a in animals: # vòng lặp for duyệt qua từng phần tử, mỗi lần lặp giá trị của phần tử được gán vào biến a
    count+=1
    print(" Số phần tử trong danh sách là:", count) # sau khi vòng lặp kết thúc dòng này sẽ in ra màn hình chuỗi và giá trị cuối cùng của biến count
```
