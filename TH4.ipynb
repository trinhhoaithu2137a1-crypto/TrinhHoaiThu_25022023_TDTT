```python
#W4A2
import math as mt
try:
    n = int(input("Nhập vào số nguyên dương n: "))
    for _ in range(100):
        if n >= 0:
            print(n,"là số nguyên dương")
            break
    uoc = 0
    for i in range(1, int(mt.sqrt(n))+1):
         du = n%i
         if du == 0:
             uoc += 1
             if i*i != n:
                 uoc +=1
    if uoc == 2:
        print(n, "là số nguyên tố")
    else:
        print(n, "không phải là số nguyên tố")
except ValueError:
    print("Kí tự bạn nhập không phải số nguyên dương")
```

```python
#W4A6
import math as mt
try:
    for _ in range(10):
        a, b = map(int, input().split())
        if a <= 0 or b <= 0:
            print("Bạn không nhập số nguyên dương")
        else:
            sum_digit = 0
            for i in range(a, b+1):
                count = 0
                for x in range(1, int(mt.sqrt(i)+1)):
                    du = i%x
                    if du == 0:
                        count += 1
                        if x*x != i:
                            count += 1
                if count == 2:
                    sum_digit += i
            print("Tổng các số nguyên tố là: ", sum_digit)
except ValueError:
    print("Bạn không nhập hai số nguyên dương")
```

```python
#W4A7
try:
    n = int(input("Nhập vào số nguyên dương n: "))
    if n<=0:
        print("Bạn đã nhập số nguyên không dương")
    elif n==1:
        print("1 không có ước nguyên tố")
    else:
        uoc_ngto_max = 0
        for i in range(1, n+1):
           du = n%i
           if du == 0:
              count = 0
              for x in range(1, i+1):
                 if i%x == 0:
                    count += 1
              if count == 2:
                 uoc_ngto_max = i
        print(f"Ước nguyên tố lớn nhất của {n} là: {uoc_ngto_max}")
except ValueError:
    print("Bạn không nhập số nguyên dương")
```

```python
#W4A8
try:
    n = int(input("Nhập vào số nguyên dương n: "))
    steps = 0
    if n <= 0:
        print("Bạn đã nhập số nguyên không dương")
    else:
        for i in range(1000):
            s = str(n)
            if s == s[::-1]:
                steps = i
                break
            reversed_s = s[::-1]
            reversed_n = int(reversed_s)
            n = n + reversed_n
        print(f"Số bước là: {steps}\nGiá trị đối xứng là: {n}")
except ValueError:
    print("Bạn không nhập số nguyên dương n")
```

```python
#W4A9
import math as mt
try:
    n = int(input("Nhập vào số nguyên dương n: "))
    if n <= 0:
        print("Bạn đã nhập số nguyên không dương")
    else:
        for i in range(1, n+1):
            if mt.sqrt(i) == int(mt.sqrt(i)) and len(set(str(i))) == len(str(i)):
                print(i, end=' ')
except ValueError:
    print("Bạn không nhập số nguyên dương")
```

```python
#W4A13
a = int(input("Nhập vào số nguyên dương thứ nhất: "))
b = int(input("Nhập vào số nguyên dương thứ hai: "))
sum_digit_1 = 0
sum_digit_2 = 0
if a<=0 or b<=0:
    print("Nhập vào hai số nguyên dương")
else:
    for i in range(1, a):
        if a%i == 0:
            sum_digit_1 += i
    for i in range(1, b):
        if b%i == 0:
            sum_digit_2 += i
    if sum_digit_1==b and sum_digit_2==a:
        print("True")
    else:
        print("False")
```
