```python
#Bài 4
try:
    h = int(input("Nhập vào h:"))
    if h==1:
        print("*")
    else:
        for i in range(1, h):
            so_luong_khoang_trang_1 = h - i 
            so_luong_khoang_trang_2 = 2*i-3
            khoang_trang = " "
            dau_sao = "*"
            khoang_trang_1 = " " * so_luong_khoang_trang_1
            khoang_trang_2 = " " * so_luong_khoang_trang_2
            hang_cuoi = (2*h-1)*dau_sao
            if i==1:
                print((h-1)*khoang_trang+dau_sao)
            else:
                print(khoang_trang_1 + dau_sao + khoang_trang_2+dau_sao)
        print(hang_cuoi)
except ValueError:
    print("Bạn không nhập số nguyên")
```

```python
#Bài 9 
import math as mt
try:
    a = int(input("Nhập vào số nguyên dương a:"))
    if a<=0:
        print("Bạn không nhập số nguyên dương")
    else:
        if mt.sqrt(a)==int(mt.sqrt(a)):
            print(f"{a} là số chính phương")
        else:
            print(f"{a} không phải là số chính phương")
except ValueError:
    print("Bạn không nhập số nguyên")
```

```python
#Bài 16
try:
    n = int(input("Nhập vào số nguyên dương n: "))
    chan = 0
    le = 0
    if n<=0:
        print("Bạn đã không nhập một số nguyên dương")
    else:
        for i in range(len(str(n))):
            du = n % 10
            if du%2==0:
                chan+=1
            else:
                le+=1
            n//=10
        print("Số chữ số chẵn là:",chan, "\nSố chữ số lẻ là:",le)
except ValueError:
    print("Bạn không nhập một số nguyên")
```

```python
#Bài 18
try:
    n = int(input("Nhập vào số nguyên dương n:"))
    n_sau = n
    k = 0
    if n<=0:
        print("Bạn không nhập vào số nguyên dương")
    else:
        if n == 1:
            print("Số 1 có dạng 3^0")
        else:
            for i in range(1,10000):
                if n_sau % 3 == 0:
                    k = i
                    n_sau /=3
                    if n_sau == 1:
                        print(f"Số {n} có dạng 3^{k}")
                        break
                else:
                    print(f"Số {n} không có dạng 3^k")
                    break
except ValueError:
    print("Bạn không nhập một số nguyên")
```

```python
#Bài 19
try:
    A = int(input("Nhập vào số nguyên dương A:"))
    if A<=0:
        print("Bạn không nhập một số nguyên dương")
    else:
        a = 1
        b = 1
        for i in range(10000):
            c = a+b
            b = a 
            a = c
            if c>=A:
               break
        print(f"Số fibonacci lớn nhất nhưng không vượt qua {A} là: {b}")
except ValueError:
    print("Bạn không nhập một số nguyên")
```

```python
#Bài 20:
try:
    n = int(input("Nhập vào số nguyên dương n:"))
    n_moi=n
    dem = 0
    if n<=0:
        print("Bạn không nhập vào một số nguyên dương")
    else:
        for i in range(10000):
            du=n_moi%10
            n_moi//=10
            if du==2 or du==3 or du==5 or du==7:
                dem+=1
            if n_moi//10==0:
                break
        print(f"Số chữ số nguyên tố trong số {n} là: {dem}")
except ValueError:
    print("Bạn không nhập một số nguyên")
```
