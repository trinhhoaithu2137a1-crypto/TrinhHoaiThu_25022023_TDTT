```python
#Bài 3
import math as mt
epsilon = 0.0000000001
a, b, c = [float(eval(x)) for x in input("Nhập 3 cạnh của tam giác:").split()]
a2= a**2
b2= b**2
c2= c**2
p=(a+b+c)/2
chuvi= a+b+c
s=mt.sqrt(p*(p-a)*(p-b)*(p-c))
is_can = (abs(a-c)<epsilon) or (abs(a-b)<epsilon) or (abs(b-c)<epsilon)
is_vuong = (abs(a2+b2-c2)<epsilon) or (abs(a2+c2-b2)<epsilon) or (abs(b2+c2-a2))
if a+b>c and a+c>b and b+c>a and a>0 and b>0 and c>0:
    if a==b and b==c:
        print ("Đây là tam giác đều")
    elif is_can and is_vuong:
        print ("Đây là tam giác vuông cân")
    elif is_vuong:
        print ("Đây là tam giác vuông")
    elif is_can:
        print ("Đây là tam giác cân")
    else :
        print ("Đây là tam giác thường")
    print("Diện tích:",s,"Chu vi:",chuvi)
else:
    print ("Không là 3 cạnh của tam giác")
```

```python
#Bài 5
print ("Giải phương trình bậc 2: ax^2 + bx + c = 0")
import math as mt
a = float(input("Nhập hệ số a:"))
b = float(input("Nhập hệ số b:"))
c = float(input("Nhập hệ số c:"))
delta = b**2 - 4*a*c
if a == 0:
    if b == 0:
        if c == 0:
            print ("Phương trình vô số nghiệm")
        else:
            print ("Phương trình vô nghiệm")
    else:
        x = -c/b
        print ("Nghiệm của phương trình là x=",x)
else:
    if delta < 0:
        print ("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b/2*a
        print ("Phương trình có nghiệm kép x=",x)
    elif delta > 0:
        x1 = (-b - mt.sqrt(delta))/2*a
        x2 = (-b + mt.sqrt(delta))/2*a
        print ("Phương trình có 2 nghiệm x1=",x1,", x2=",x2)
        
```

```python
#Bài 8
print("---Giải hệ phương trình tuyến tính---")
print("         ax + by = m")
print("         cx + dy = n")
a = float(input("Nhập vào hệ số a:"))
b = float(input("Nhập vào hệ số b:"))
m = float(input("Nhập giá trị của m:"))
c = float(input("Nhập vào hệ số c:"))
d = float(input("Nhập vào hệ số d:"))
n = float(input("Nhập giá trị của n:"))
if a/c == b/d and a/c != m/n:
    print("Hệ phương trình vô nghiệm")
elif a/c == b/d and b/c == m/n:
    print("Hệ phương trình vô số nghiệm")
else:
    x = (b*n-d*m)/(c*b-d*a)
    y = (a*n-c*m)/(a*d-c*b)
    print(f"Nghiệm của hệ phương trình là:(x,y)=({x},{y})")
```

```python
#Bài 9
print("Chuyển đổi thời gian sang giây")
x = float(input("Nhập thời gian công việc:"))
if x < 60:
    print(f"Thời gian là :0 giờ, 0 phút, {x} giây")
elif 60 < x < 3600:
    phut = x // 60
    so_giay_du = x % 60
    print(f"Thời gian là :0 giờ, {phut} phút, {so_giay_du} giây")
else:
    gio = x // 3600
    so_giay_du = x % 3600
    if so_giay_du < 60:
        print(f"Thời gian là :{gio} giờ, 0 phút, {so_giay_du} giây")
    else:
        phut = so_giay_du // 60
        so_giay_du = so_giay_du % 60
        print(f"Thời gian là :{gio} giờ, {phut} phút, {so_giay_du} giây")
```

```python
#Bài 10 
print("---Xác định điểm có thuộc đường tròn---")
x_r = float(input("Nhập vào hoành độ tâm I:"))
y_r = float(input("Nhập vào tung độ tâm I:"))
r = float(input("Nhập vào bán kính hình tròn:"))
if r <= 0:
    print("Không tồn tại hình tròn")
else:
    x_A = float(input("Nhập vào hoành độ điểm A:"))
    y_A = float(input("Nhập vào tung độ điểm A"))
    if (x_A - x_r)**2 + (y_A-y_r)**2 == r**2:
        print("Điểm A thuộc đường tròn")
    else:
        print("Điểm A không thuộc đường tròn")
```
